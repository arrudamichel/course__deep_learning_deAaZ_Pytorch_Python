```python
import torch
import torch.nn as nn
from torch import optim
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from torchvision import datasets
import torchvision.transforms as transforms
import torchvision
import zipfile
torch.__version__
```

## 3. Construção do modelo

```python
# output = (input - filter + 1) / stride (=1)
# stride -> de quantos em quantos pixels é o janelamento para convolucao
classificador = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3),
            nn.ReLU(),
            nn.BatchNorm2d(num_features=32),
            # (64 - 3 + 1) / 1 = 62x62
            nn.MaxPool2d(kernel_size=2),
            # 31 x 31 (metade por causa do 2)
            nn.Conv2d(in_channels=32, out_channels=32, kernel_size=3),
            nn.ReLU(),
            nn.BatchNorm2d(num_features=32),
            # (31 - 3 + 1) / 1 = 29x29
            nn.MaxPool2d(kernel_size=2),
            # 14x14
            nn.Flatten(),
            # 6272 -> 128 -> 128 -> 1
            # 14 * 14 * 32 -> saida MaxPool2d * saida MaxPool2d * numero de filtros
            nn.Linear(in_features=14*14*32, out_features=128),
            nn.ReLU(),
            nn.Dropout(p=0.2),
            nn.Linear(in_features=128, out_features=128),
            nn.ReLU(),
            nn.Dropout(p=0.2),
            nn.Linear(in_features=128, out_features=1),
            nn.Sigmoid()
)
```

```python
criterion = nn.BCELoss()
optimizer = optim.Adam(classificador.parameters())
```

## 2. Importando dados

```python
torch.manual_seed(123)
```

```python
data_dir_train = 'data/dataset/training_set'
data_dir_test = 'data/dataset/test_set'
```

```python
transform_train = transforms.Compose([
                                      transforms.Resize([64, 64]),
                                      transforms.RandomHorizontalFlip(),
                                      transforms.RandomAffine(degrees = 7, 
                                                              translate = (0, 0.07), # horizontal and vertical shifts
                                                              shear = 0.2,
                                                              scale = (1, 1.2) # zoom range
                                                              ),
                                      transforms.ToTensor()
                                      ])
```

```python
transform_test = transforms.Compose([
                                      transforms.Resize([64, 64]),
                                      transforms.ToTensor()
                                      ])
```

```python
train_dataset = datasets.ImageFolder(data_dir_train, transform=transform_train)
```

```python
train_dataset
```

```python
test_dataset = datasets.ImageFolder(data_dir_test, transform=transform_test)
```

```python
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True, pin_memory=False,  # MPS não suporta pin_memory=True
                                                                              generator=torch.Generator(device=device))

                                                                            
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=True, pin_memory=False,  # MPS não suporta pin_memory=True
                                                                              generator=torch.Generator(device=device))
```

## 4. Treinamento do modelo

```python
classificador.to(device)
```

```python
def training_loop(loader, epoch):
    running_loss = 0.
    running_accuracy = 0.

    for i, data in enumerate(loader):
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = classificador(inputs)

        loss = criterion(outputs, labels.float().view(*outputs.shape))
        loss.backward()

        optimizer.step()

        running_loss += loss.item()

        predicted = torch.tensor([1 if output > 0.5 else 0 for output in outputs]).to(device)

        equals = predicted == labels.view(*predicted.shape)

        accuracy = torch.mean(equals.float())
        running_accuracy += accuracy

        print('\rEpoch: {:4d} - Loop {:3d} of {:3d} - Loss: {:6.4f} - Accuracy: {:6.4f}'.format(epoch + 1,
                      i + 1,
                      len(loader),
                      loss,                                                                          
                      accuracy), end = '\r')

    print(f"\rEnded Epoch {epoch + 1} - loss: {running_loss/len(loader)} - precision: {running_accuracy/len(loader)} end=\r")
```

```python
for epoch in range(10):
    print("Training...")
    training_loop(train_loader, epoch)
    
    print("Validating...")
    classificador.eval()
    training_loop(test_loader, epoch)
    classificador.train()
```

```python
def classificar_imagem(fname):
    from PIL import Image
    import matplotlib.pyplot as plt
    imagem_teste = Image.open(data_dir_test + '/' + fname)
    plt.imshow(imagem_teste)

    import numpy as np
    imagem_teste = imagem_teste.resize((64, 64)) #padroniza tamanho da imagem em 64 x 64
    imagem_teste = np.array(imagem_teste.getdata()).reshape(*imagem_teste.size, 3) # converte para np array
    imagem_teste = imagem_teste / 255 # normalizacao
    imagem_teste = imagem_teste.transpose(2, 0, 1) #batch_size primeiro e canais no final
    imagem_teste = torch.tensor(imagem_teste, dtype=torch.float).view(-1, *imagem_teste.shape)

    classificador.eval()
    imagem_teste = imagem_teste.to(device)
    output = classificador.forward(imagem_teste)
    if output > 0.5:
        output = 1
    else:
        output = 0

    print("Previsão: ", output)

    idx_to_class = { value : key for key, value in test_dataset.class_to_idx.items() }

    return idx_to_class[output]
    
```
