## 3. Constru√ß√£o dos modelos

```python
class regressor_torch(nn.Module):
    def __init__(self):
        super().__init__()
        # (99 + 1) / 2 = 50
        
        self.dense0 = nn.Linear(99, 50)

        self.dense1 = nn.Linear(50, 50)

        self.activation = nn.Sigmoid()
        self.output = nn.Linear(50, 1)

    def forward(self, X):
        X = self.dense0(X)
        X = self.activation(X)
        X = self.dense1(X)
        X = self.activation(X)
        X = self.output(X)
        return X
```

```python
regressor = regressor_torch()
```

```python
criterion = nn.MSELoss()
optimizer = optim.Adam(regressor.parameters())
```

```python
type(previsores)
```

```python
previsores = torch.tensor(previsores, dtype=torch.float)
vendas = torch.tensor(np.array(vendas), dtype=torch.float)
```

```python
dataset = torch.utils.data.TensorDataset(previsores, vendas)
train_loader = torch.utils.data.DataLoader(dataset, 
                                           batch_size=10,
                                           shuffle=True,
                                           pin_memory=False,  # MPS n√£o suporta pin_memory=True
                                           generator=torch.Generator(device=device))
```

```python
regressor.to(device)
```

## 4. Treinamento

```python
for epoch in range(2000):
    
    running_loss = 0.
    
    for i, data in enumerate(train_loader):
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)
       
        optimizer.zero_grad()

        outputs = regressor(inputs)
        
        loss = criterion(outputs, labels)
        loss.backward()
        
        optimizer.step()

        loss = loss.item()
        running_loss += loss * len(inputs)

        print('\rEpoch {:4d} - Loop {:3d} of {:3d} - cost {:6.4f}'.format(epoch + 1,
                      i + 1,
                      len(train_loader),
                      loss), end = '\r')
        
    print('Final Epoch {:4d} - cost {:6.4f}         '.format(epoch + 1, 
         running_loss/len(dataset)))
        
```
